```python
#A01793599 - Initial part of the lab

import pandas as pd
import matplotlib.pylab as plt
import numpy as np


filename="auto.csv"

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]


df = pd.read_csv(filename, names = headers)

df.head()
```

Claramente se nota que hay valores faltantes, ahora a complementar los datos

```python
#A01793599 - QUESTION 1 - Cambiar NaN por Promedio

# A01793599 - Primero homologamos los valores a Nan
df.replace("?", np.nan, inplace = True)

# A01793599 - Calculando mean value - Promedio
meanValue = df["stroke"].astype("float").mean( axis = 0)
print( "Promedio: ", meanValue)

# A01793599 - Utilizando replace para cambiar NaN por Mean Value
df["stroke"].replace(np.nan, meanValue, inplace=True)

df.head()
```

```python
#A01793599 - QUESTION 2 - Convertir MPG a L/100km en la columna "highway-mpg", renombrar a "highway-L/100km

# A01793599 - Primero convertimos diviendo 235 entre el total de millas Highway-mpb
df["highway-mpg"] = 235 / df["highway-mpg"]

# A01793599 - Renombramos las columnas
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)

df.head()
```

```python
#A01793599 - QUESTION 3 - Normalizar la columna Height

# A01793599 - Realizamos la normalización utilizando el método de Simple feature Scaling Xnew = Xold / Xmax
df['height'] = df['height'] / df['height'].max() 

# A01793599 - Mostramos la columna normalizada
df[["height"]].head()
```

```python
#A01793599 - QUESTION 4 - Crear una variable para la columna "aspiration"

# A01793599 - Obtenemos la conversión de valores categóricos hacia numéricos
valoresNumericos = pd.get_dummies(df['aspiration']) 

# A01793599 - Dividimos en dos columnas por claridad 
valoresNumericos.rename(columns={'std':'valor-std', 'turbo': 'valor-turbo'}, inplace=True)

# A01793599 - Mostramos la variable numérica
valoresNumericos.head(10)
```

```python
#A01793599 - QUESTION 5 - Combina la nueva variable con el DataFrame original borrando la columna "aspiration"

# A01793599 - Concatenamos los dataframes
df = pd.concat([df, valoresNumericos], axis=1)

# A01793599 - Eliminamos la columna original manteniendo un solo dataframe DF 
df.drop('aspiration', axis = 1, inplace=True)

# A01793599 - Mostramos el dataframe
df.head(10)
```
